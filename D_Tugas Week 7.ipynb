```python
#Buatlah sebuah fungsi untuk mencari invers matrix berdasarkan input matrix dari user

def invers():
    det = []

    #Tulis kode kamu disini
def invers():
    
    baris = int(input('Masukkan Jumlah Baris: '))
    kolom = int(input('Masukkan Jumlah Kolom: '))
    
    matrix = create_matrix(baris, kolom)
    det = determinan(matrix)

    if det == 0:
        print("Matriks tidak memiliki invers karena determinan sama dengan 0.")
    else:
        inverse_matrix = inverse(matrix, det)
        print()
        print_matrix(inverse_matrix)


def create_matrix(baris, kolom):
    matrix = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            element = float(input(f'Masukkan elemen matriks pada baris ke-{i+1}, kolom ke-{j+1}: '))
            row.append(element)
        matrix.append(row)
    return matrix


def determinan(matrix):
    n = len(matrix)
    if n != len(matrix[0]):
        raise ValueError
    
    if n == 1:
        return matrix[0][0]
    
    det = 0
    for j in range(n):
        submatrix = []
        for i in range(1, n):
            row = matrix[i][:j] + matrix[i][j+1:]
            submatrix.append(row)
        det += (-1) ** j * matrix[0][j] * determinan(submatrix)
    
    return det


def cofactor(matrix):
    n = len(matrix)
    cofactor_matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            submatrix = []
            for k in range(n):
                if k != i:
                    subrow = matrix[k][:j] + matrix[k][j+1:]
                    submatrix.append(subrow)
            sign = (-1) ** (i + j)
            cofactor_ij = sign * determinan(submatrix)
            row.append(cofactor_ij)
        cofactor_matrix.append(row)
    return cofactor_matrix


def transpose(matrix):
    n = len(matrix)
    transposed_matrix = []
    for j in range(n):
        row = []
        for i in range(n):
            row.append(matrix[i][j])
        transposed_matrix.append(row)
    return transposed_matrix


def inverse(matrix, det):
    n = len(matrix)
    cofactor_matrix = cofactor(matrix)
    adjugate_matrix = transpose(cofactor_matrix)
    inverse_matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            element = adjugate_matrix[i][j] / det
            row.append(element)
        inverse_matrix.append(row)
    return inverse_matrix


def print_matrix(matrix):
    for row in matrix:
        print(row)


invers()

print()
```
